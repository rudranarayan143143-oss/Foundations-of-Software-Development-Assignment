#Question 4:
You are setting up a web server on a Linux machine. You run your server script but it says 'Permission denied'. Your senior tells you the file needs to be executable. What will you do to fix this?
#Answer:
To make a file executable you need to use the `chmod` command. We do this by using the `chmod` command like this:
chmod +x your_script.sh
The thing is, in Linux files have permissions that say what we can do with them. We can read them write to them or run them. Lots of scripts are not allowed to be run by default because of security.
When we use the `+x` flag with the `chmod` command you are telling the system to let the file be run like a program or script. The `chmod` command is what makes this happen. We use the `chmod` command to change the permissions of the file. The `+x` flag is important because it gives the file permission to be executed. So the `chmod` command, with the `+x` flag is what makes our script executable.

#Question 7:
A company's website is not working. The developer checks the HTML file and sees
there is no DOCTYPE declaration at the top and the content is placed directly without any head or body tags. Why could this be a problem?
#Answer:
Without a DOCTYPE the browser thinks the website is from the 1990s. It switches to Quirks Mode, which makes your colors and boxes and fonts look really messy and wrong.
The head is like the instruction manual, for the website. Without it the browser does not know the pages name or how to load your styling, which's the CSS.
The body tells the browser where the actual content of the website starts. Without it your code is a pile of text, which makes it hard for Google or screen readers to understand the website.
Interactive parts of the website, which are the JavaScript parts will crash because they are looking for a body tag that is not there.
Basically the browser has to guess how to show the website and the browser usually guesses wrong.

#Question 8:
You are building an online store page for a shop in your city. The page has a product listing, a contact form, a header with the shop name, and a footer with address. Your senior says you should use semantic HTML tags. Which tags will you use for each section and why?
#Answer:
When you use tags you are telling the browser and search engines what each part of your code is really about, not just what it looks like. So here is how you can set up your store.
You can start with the <header> tag for the shop name. This tag is for the section of your store. It helps to identify the content and navigation for the page.
Next you can use the <> tag as the wrapper. You should put everything between the header and footer inside this tag. The <main> tag tells the browser where the unique and primary content of your page is located.
Then you can use the <section> tag for the product listing. This tag is used to group your products. It defines an area of the page that is about one theme.
For each product you can use the <article> tag. Each product should be an article. This tells the browser that the product information, like the image and price and name of the product is a self-contained unit that can make sense even if it is shared on websites.
If you have a contact form you can use the <form> tag. This is the semantic tag for user input. It makes sure the browser handles the data submission correctly.
Finally you can use the <footer> tag for the address. This tag is for the section of your store. It is designed to hold information, like contact details and copyright and addresses which are typically found at the end of a page.

#Question 9:
What is the difference between > and >> in the terminal? Give one example of
when you would use each.
#Answer:
">(Overwrite)"

This takes the output of a command and writes it to a file. If the file already has text in it, it eraseseverything and replaces it with the new output.

If you want to save a list of your current files to a fresh file called inventory.txt, you run: ls > inventory.txt


">> (Append)"

This takes the output and adds it to the end of the file. It keeps the existing text safe and just grows the file.

If you are tracking errors over time and want to add a new error message to your logs.txt without deleting the old ones, you run: echo "Error at 10:00 AM" >> logs.txt

#Question 10:
What is the difference between rm and rm -r? Why should you be careful when using rm -r?
#Answer:
"rm (Remove):"

This is used to delete individual files. If you try to use it on a folder (directory), the terminal will give you an error and refuse to delete it.

Example: rm notes.txt

"rm -r (Recursive)"

The -r stands for recursive. This tells the terminal to go inside a folder and delete everything—every subfolder, every file, and the folder itself.

Example: rm -r myapp

Unlike dragging a file to the "Trash" or "Recycle Bin" on your desktop, there is no undo in the terminal. Once you run this command, those files are gone.
